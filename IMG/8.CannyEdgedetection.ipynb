Canny Edge Detection

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Load Image (Grayscale)
img = cv2.imread("art.png", cv2.IMREAD_GRAYSCALE)

# Step 2: Gaussian Blur
blur = cv2.GaussianBlur(img, (5,5), 5.4)

# Step 3: Canny Edge Detection
edges = cv2.Canny(blur, 100, 100)

# Step 4: Show Results
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')

plt.subplot(1,3,2)
plt.title("Blurred")
plt.imshow(blur, cmap='gray')

plt.subplot(1,3,3)
plt.title("Canny Edges")
plt.imshow(edges, cmap='gray')

plt.show()
cv2.imwrite("8.artCanny.png"  , figure)
```
