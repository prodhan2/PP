✅ 1. Huffman Coding (Using built-in cv2.imencode & cv2.imdecode)

```python
import cv2
import matplotlib.pyplot as plt

# Load original image
original = cv2.imread("nature3.png", 0)  # grayscale

# Huffman / JPEG compression
cv2.imwrite("nature3_huffman.jpg", original, [cv2.IMWRITE_JPEG_QUALITY, 50])

# cv2.imwrite("nature3_png3.png", original, [int(cv2.IMWRITE_PNG_COMPRESSION), 3])
huffman = cv2.imread("nature3_huffman.jpg", 0)

# Plot original + compressed
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(original, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(huffman, cmap='gray')
plt.title("Huffman / JPEG")
plt.axis('off')
plt.tight_layout()


print("PART 1 DONE → Saved as nature3_huffman.jpg")
```

⭐ PART 2 — DCT Compression (Show + Save)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load original image
original = cv2.imread("nature3.png", cv2.IMREAD_GRAYSCALE)
img_f = np.float32(original)

# DCT compression
dct = cv2.dct(img_f)
dct[np.abs(dct) < 20] = 0  # Thresholding
compressed = cv2.idct(dct)
compressed = np.uint8(np.clip(compressed, 0, 255))

# Save compressed
cv2.imwrite("nature3_dct.png", compressed)

# Plot original + compressed
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(original, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(compressed, cmap='gray')
plt.title("DCT Compressed")
plt.axis('off')
plt.tight_layout()
plt.show()

print("DCT Compression Done → nature3_dct.png")
```

⭐ PART 3 — DWT Compression (Show + Save)

```python
import cv2
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Load original image
original = cv2.imread("nature3.png", cv2.IMREAD_GRAYSCALE)

# DWT compression
LL, (LH, HL, HH) = pywt.dwt2(original, 'haar')
LH = pywt.threshold(LH, 20, mode='soft')
HL = pywt.threshold(HL, 20, mode='soft')
HH = pywt.threshold(HH, 20, mode='soft')
compressed = pywt.idwt2((LL, (LH, HL, HH)), 'haar')
compressed = np.clip(compressed, 0, 255).astype(np.uint8)

# compressed = np.uint8(np.clip(pywt.idwt2((LL, (LH, HL, HH)), 'haar'), 0, 255))


# Save compressed
cv2.imwrite("nature3_dwt.png", compressed)

# Plot original + compressed
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(original, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(compressed, cmap='gray')
plt.title("DWT Compressed")
plt.axis('off')
plt.tight_layout()
plt.show()

print("DWT Compression Done → nature3_dwt.png")
```

CODE FOR JPEG AND PNG Compression

```python
import cv2
import matplotlib.pyplot as plt

# Load original image
img = cv2.imread("nature3.png", 0)  # grayscale

# =========================
# JPEG Compression
# =========================
cv2.imwrite('nature3_jpeg90.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 1])
jpeg_img = cv2.imread('nature3_jpeg90.jpg', 0)

# =========================
# PNG Compression
# =========================
cv2.imwrite('nature3_png3.png', img, [int(cv2.IMWRITE_PNG_COMPRESSION), 9])
png_img = cv2.imread('nature3_png3.png', 0)

# =========================
# Plot Original + Compressed
# =========================
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(jpeg_img, cmap='gray')
plt.title("JPEG Compressed (Q=90)")


plt.subplot(1,3,3)
plt.imshow(png_img, cmap='gray')
plt.title("PNG Compressed (Level=3)")


plt.tight_layout()
plt.show()

print("All images compressed and saved successfully!")
```
